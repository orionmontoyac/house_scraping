```python
la_haus_medellin = pd.read_csv('medellin_la_haus_2020_03_19_houses.csv')
la_haus_envigado = pd.read_csv('envigado_la_haus_2020_03_20_houses.csv')
la_haus_sabaneta = pd.read_csv('sabaneta_la_haus_2020_03_20_houses.csv')

la_haus = pd.concat([la_haus_medellin, la_haus_envigado, la_haus_sabaneta])
la_haus = la_haus.reset_index()
la_haus
```

```python
houses = pd.concat([la_haus, finca_raiz])
houses = houses.reset_index()
houses
```

```python
# Añadir house uid al DataFrame
from urllib.parse import urlparse
houses['house_uid'] = houses['link'].apply(lambda link: urlparse(link).hostname).apply(lambda host: host.split('.')[1])
houses
```

```python
# Obtener el host
houses['host'] = houses['link'].apply(lambda link: urlparse(link).hostname)
houses = houses.drop(columns=['level_0', 'index'])
houses
```

```python
# Añadiendo precio del metro cuadrado m2_price
houses['m2_price'] = houses.apply(lambda x: round(x.price/x.area,2), axis=1)
houses
```
