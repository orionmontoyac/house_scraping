```python
import pandas as pd
```

```python
la_haus_medellin = pd.read_csv('medellin_la_haus_2020_03_19_houses.csv')
la_haus_envigado = pd.read_csv('envigado_la_haus_2020_03_20_houses.csv')
la_haus_sabaneta = pd.read_csv('sabaneta_la_haus_2020_03_20_houses.csv')

la_haus = pd.concat([la_haus_medellin, la_haus_envigado, la_haus_sabaneta])
la_haus = la_haus.reset_index()
la_haus
```

```python
finca_raiz = pd.read_csv('antioquia_finca_raiz_2020_03_20_houses.csv')
finca_raiz
```

```python
houses = pd.concat([la_haus, finca_raiz])
houses = houses.reset_index()
houses
# Eliminando casas con precios menores a 200 millones
houses = houses[houses['price'] >= 200]
```

```python
houses = houses.drop(['index','level_0'], axis = 1)
houses
```

```python
# Añadir house uid al DataFrame
from urllib.parse import urlparse
houses['house_uid'] = houses['link'].apply(lambda link: urlparse(link).hostname).apply(lambda host: host.split('.')[1])
houses
```

```python
# Añadiendo precio del metro cuadrado m2_price
houses['m2_price'] = houses.apply(lambda x: round(x.price/x.area,2), axis=1)
houses.head()
```

```python
# Quitando acentos y mayusculas en texto de zone
import unidecode
houses['zone'] = houses['zone'].apply(lambda x: unidecode.unidecode(x).lower())
houses.head()
```

```python
# Elimnando casas que no esten en Medellin, Sabaneta o Envigado
from fuzzywuzzy import fuzz

no_excepts = ['medellin', 'sabaneta','envigado']

def check_except(text):
    for no_except in no_excepts:
        simil = fuzz.token_set_ratio(text, no_except)
        if simil == 100:
            return False
        
    return True

houses = houses[houses['zone'].apply(check_except) == 0]
houses.reset_index()
```

```python
import plotly.graph_objects as go

def group_values(pd_series, gap):
    group_values = {}
    index = pd_series
    max_value = int(max(index))
    min_value = int(min(index))
   
    for ran in range(min_value, max_value, gap):
        #print(ran,'-',ran+gap,len([price for price in index if price >= ran and price < ran + gap]))
        group_values[str(ran) + '-' + str(ran+gap)] = len([price for price in index if price >= ran and price < ran + gap])
    #print(group_values.keys())  
    
    return list(group_values.keys()), list(group_values.values())


price_histo = houses['price'].value_counts()

labels, values = group_values(houses['price'], 10)
```
